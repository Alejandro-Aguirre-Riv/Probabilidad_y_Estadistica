# **Diagramas de cajas y bigotes**

```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2 )
    fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(40, 93), 2 )
    fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
plt.boxplot(fabrica1)
plt.ylabel('Produccion en mts.')
plt.title('Produccion en la fabrica 1')
plt.show()
```

```python
plt.boxplot(fabrica2)
plt.ylabel('Produccion en mts.')
plt.title('Produccion en la fabrica 2')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1,
                    vert=True,
                    patch_artist=True)
ax1.set_title('Produccion Fabrica 1')

bplot2 = ax2.boxplot(fabrica2,
                     vert=True,
                    patch_artist=True)

ax2.set_title('Produccion Fabrica 2')
plt.show()
```

```python
fabrica1.append(150)
print(fabrica1)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1,
                    vert=True,
                    patch_artist=True)
ax1.set_title('Produccion Fabrica 1')

bplot2 = ax2.boxplot(fabrica2,
                     vert=True,
                    patch_artist=True)

ax2.set_title('Produccion Fabrica 2')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 5))
bplot1 = ax1.boxplot(fabrica1,
                    vert=True,
                    patch_artist=True)
ax1.set_title('Produccion Fabrica 1')
ax1.set_ylim(0, 160) #escala uniforme

bplot2 = ax2.boxplot(fabrica2,
                     vert=True,
                    patch_artist=True)

ax2.set_title('Produccion Fabrica 2')
ax2.set_ylim(0, 160)
plt.show()
```
